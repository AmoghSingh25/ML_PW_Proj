```python
from simulation import simulation
import matplotlib.pyplot as plt
```

```python
sim = simulation(
    wave_freq=20,
        wave_speed=1,
        wave_decay=0.6,
        wave_cutoff=0.1,
        wave_retreat_coeff=0.8,
        wave_height=2,
        sand_pull=0.2,
        ground_pull=0.1,
        water_decay=0.99,
        wave_vol=1,
        wave_amplitude=50,
        wave_spread=0
)
ret = sim.run_sim(100)
```

```python
coords1 = sim.get_coast_coords(ret[0], limit=0)
coords2 = sim.get_coast_coords(ret[1], limit=0)
sim.plot_coast(coords1, label='Before erosion')
sim.plot_coast(coords2, label='After erosion')
plt.legend()
plt.show()
```
